```python
#Считываем данные в массив:

import pandas
df = pandas.read_csv('./dataset_Facebook.csv', sep = ';')
df = df.dropna(0, how = 'any')
```

```python
df.head()
```

```python
#Среднее значение для каждого стобца:

df =df.drop(columns=['Type'])
mean_cols = {column: df[column].mean() for column in df}
print mean_cols 
```

```python
#Среднее для всего массива:
mean_all= 0
for key, value in mean_cols.iteritems():
    mean_all+= value
    
print mean_all/len(mean_cols)
```

```python
#Максимальное/минимальное для каждого столбца:

df.describe()
```

```python
#Для каждого типа отдельно:

df.groupby('Type').describe().unstack(level=0)
```

```python
#Найдем моду для каждого типа:
def mod_table(df) :
    modes = df.mode()
    return modes
from scipy import stats
modes = pandas.DataFrame(df.groupby('Type').apply(mod_table)).fillna("")
df.groupby(['Type']).agg(lambda x: stats.mode(x)[0])
```

```python
#Найдем самый популярный объяект в выборке:
#На мой взгляд надо искать наиболее высокую плотность активности за период времени:
from itertools import count, izip

popular = df['Total Interactions'].div(df['Lifetime Post Total Reach'], axis=0)


df.iloc[max(izip(popular, count()))[1]]
```
